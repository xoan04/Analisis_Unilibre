```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Generar el DataFrame aleatorio
random_data = np.random.rand(100, 8)
columns = [f"Columna_{i}" for i in range(1, 9)]
df_random = pd.DataFrame(random_data, columns=columns)

# Mostrar el resumen estadístico
stats = df_random.describe()
print(stats)

# Graficar el resumen estadístico
plt.figure(figsize=(12, 6))
for stat in ["mean", "std", "min", "25%", "50%", "75%", "max"]:
    plt.plot(stats.columns, stats.loc[stat], marker='o', label=stat)

plt.title("Resumen Estadístico del DataFrame Aleatorio", fontsize=14)
plt.xlabel("Columnas", fontsize=12)
plt.ylabel("Valores", fontsize=12)
plt.legend(title="Estadísticas")
plt.grid(alpha=0.5)
plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt


array_1_to_10 = np.arange(1, 11)
suma = np.sum(array_1_to_10)
media = np.mean(array_1_to_10)
valor_maximo = np.max(array_1_to_10)
print("Array del 1 al 10:", array_1_to_10)
print("Suma:", suma, "| Media:", media, "| Valor máximo:", valor_maximo)


array_ceros = np.zeros((3, 4))
array_ceros_reshape = array_ceros.reshape(2, 6)
print("\nArray 3x4 lleno de ceros:\n", array_ceros)
print("\nArray cambiado a 2x6:\n", array_ceros_reshape)

# Generar una matriz aleatoria de 3x3 y calcular su transpuesta
matriz_random = np.random.rand(3, 3)
matriz_transpuesta = matriz_random.T
print("\nMatriz aleatoria 3x3:\n", matriz_random)
print("\nMatriz transpuesta:\n", matriz_transpuesta)


array_a = np.array([1, 2, 3, 4, 5])
array_b = np.array([6, 7, 8, 9, 10])
array_concatenado = np.concatenate((array_a, array_b))
print("\nArrays concatenados:\n", array_concatenado)

# Graficar los resultados del primer array para visualización
plt.figure(figsize=(10, 6))
plt.bar(range(1, 11), array_1_to_10, color='blue', alpha=0.7)
plt.title("Valores del Array del 1 al 10", fontsize=14)
plt.xlabel("Índice", fontsize=12)
plt.ylabel("Valor", fontsize=12)
plt.show()
```
