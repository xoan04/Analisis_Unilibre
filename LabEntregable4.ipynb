```python

import pandas as pd
from google.colab import files

# Parte 1: Exploración inicial del dataset
# Subir el archivo CSV
print("Por favor, sube el archivo CSV que deseas analizar y limpiar:")
uploaded = files.upload()

# Leer el archivo CSV en un DataFrame
file_name = list(uploaded.keys())[0]
df = pd.read_csv(file_name)

# Mostrar el dataset cargado
print("Dataset cargado:")
print(df.head())

# Exploración inicial
print("\nExploración inicial del dataset:")
print("Valores nulos por columna:\n", df.isnull().sum())
print("\nRegistros duplicados:", df.duplicated().sum())
if "City" in df.columns:
    print("\nConteo de valores únicos en la columna 'City':\n", df['City'].value_counts())
else:
    print("\nLa columna 'City' no existe en el dataset.")

# Parte 2: Limpieza de Datos
print("\nIniciando limpieza de datos...")

# Eliminar o reemplazar valores nulos
df.fillna(value={"City": "Unknown"}, inplace=True)  # Ejemplo para 'City'
df.dropna(inplace=True)  # Eliminar filas con valores nulos en otras columnas

# Eliminar duplicados
df.drop_duplicates(inplace=True)

# Normalización de formatos
if "City" in df.columns:
    df['City'] = df['City'].str.lower().str.strip()  # Convertir a minúsculas y eliminar espacios

# Estandarizar fechas
if "Date" in df.columns:
    # Convertir fechas al formato AAAA/MM/DD; manejar errores y valores faltantes
    df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
    # Formatear las fechas convertidas al formato estándar AAAA/MM/DD
    df['Date'] = df['Date'].dt.strftime('%Y/%m/%d')

# Reemplazar la palabra "cat" por "category" en la columna 'Category'
if "Category" in df.columns:
    df['Category'] = df['Category'].str.lower().str.strip()
    df['Category'] = df['Category'].replace(to_replace=r'\bcat\b', value='category', regex=True)

# Validación del dataset limpio
print("\nValidación después de la limpieza:")
print("Valores nulos por columna:\n", df.isnull().sum())
print("\nRegistros duplicados:", df.duplicated().sum())

# Mostrar el dataset limpio y estandarizado
print("\nDataset limpio y estandarizado:")
print(df.head())

# Exportar el dataset limpio a un nuevo archivo CSV
output_file = "dataset_limpio.csv"
df.to_csv(output_file, index=False)
print("\nEl dataset limpio se ha guardado como:", output_file)

# Descargar el archivo limpio
files.download(output_file)
```
